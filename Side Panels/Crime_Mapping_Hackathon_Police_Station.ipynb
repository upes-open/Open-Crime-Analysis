```python
police_st_csv = pd.read_csv("Police_Stations_Map.csv") # read the data
police_st_csv.drop('0', axis=1, inplace=True)
police_st_csv.head()
```

```python
criminal_csv = pd.read_csv("criminal_data.csv") # read the data
criminal_csv.head()
```

```python
df_criminal_org = pd.DataFrame(police_st_csv) # transfer to dataframe
df_criminal_org.dropna(inplace=True)
df_criminal_org
```

```python
df_criminal_org_1 = pd.DataFrame(criminal_csv[:100]) # transfer to dataframe
df_criminal_org_1.dropna(inplace=True)
df_criminal_org_1
```

```python
def popup_html_police(row):
    i = row
    police_station=df_criminal_org['Police Stations'].iloc[i]
    details = df_criminal_org['Details'].iloc[i]
    
    
    html_1 = """<!DOCTYPE html>
<html>
<head>
<h4 style="margin-bottom:10"; width="200px">{}</h4>""".format(police_station) + """
</head>
    <table style="height: 126px; width: 350px;">
<tbody>
<tr>
<td style="background-color: """+ """;"><span style="color: black;">Address & Telephone</span></td>
<td style="width: 150px;background-color: """ +""";">{}</td>""".format(details) + """
</tr>
</tbody>
</table>
</html>
"""
    return html_1
```

```python
def popup_html(row):
    i = row
    district_name=df_criminal_org_1['Police District'].iloc[i] 
    incident_datetime=df_criminal_org_1['Incident Datetime'].iloc[i] 
    report_datetime = df_criminal_org_1['Report Datetime'].iloc[i] 
    crimes = df_criminal_org_1['Crimes'].iloc[i]
    incident_description= df_criminal_org_1['Incident Description'].iloc[i]
    Resolution= df_criminal_org_1['Resolution'].iloc[i]
    Latitude = df_criminal_org_1['Latitude'].iloc[i]
    Longitude = df_criminal_org_1['Longitude'].iloc[i]

    left_col_color = "#19a7bd"
    right_col_color = "#f2f0d3"
    
    html = """<!DOCTYPE html>
<html>
<head>
<h4 style="margin-bottom:10"; width="200px">{}</h4>""".format(district_name) + """
</head>
    <table style="height: 126px; width: 350px;">
<tbody>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Report Datetime</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(report_datetime) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Incident Datetime</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(incident_datetime) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Crimes</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(crimes) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Incident description</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(incident_description) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Resolution$</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(Resolution) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Latitude $</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(Latitude) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Longitude $</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(Longitude) + """
</tr>
</tbody>
</table>
</html>
"""
    return html
```

```python
for i in range(0,len(df_criminal_org_1)):
    html = popup_html(i)
    iframe = branca.element.IFrame(html=html,width=510,height=280)
    popup = folium.Popup(folium.Html(html, script=True), max_width=500)
    folium.Marker([df_criminal_org_1['Latitude'].iloc[i],df_criminal_org_1['Longitude'].iloc[i]],
                  popup=popup,icon=folium.Icon(icon='star empty', prefix='glyphicon')).add_to(ut_map_lc) 

ut_map_lc
```

```python
for i in range(0,len(df_criminal_org)):
    html = popup_html_police(i)
    iframe = branca.element.IFrame(html=html,width=200,height=280)
    popup = folium.Popup(folium.Html(html, script=True), max_width=350)
    folium.Marker([df_criminal_org['Latitudes'].iloc[i],df_criminal_org['Longitudes'].iloc[i]],
                  popup=popup,icon=folium.Icon(icon='fa-building', prefix='fa', color = 'red')).add_to(ut_map_lc) 

ut_map_lc.save('crime_mapping_sf.html')
```

```python
address_text_box = ipywidgets.Text(value='', placeholder='type here', description='address:')

# widget function
def plot_locations(address):
    # location address
    location = geocoder.osm(address)
    
    # latitude and longitude of location
    latlng = [location.lat, location.lng]
    
    # create map
    plot_locations_map = folium.Map(location=[30.316496, 78.032188], zoom_start=12)
    
    # marker
    folium.Marker(latlng, popup=str(address), tooltip='click').add_to(plot_locations_map)
    
    # display map
    display(plot_locations_map)
    
# interaction between widget and function    
ipywidgets.interact_manual(plot_locations, address=address_text_box)

```
